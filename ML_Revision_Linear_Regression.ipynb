# Task
Analyze the California housing dataset using linear regression, with 'medianHouseValue' as the target variable. Train a linear regression model, evaluate its performance using R-squared, MAE, MSE, and RMSE, make predictions, visualize the results, and interpret the model's coefficients.

## Evaluate model performance

### Subtask:
Calculate metrics like R-squared, Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE) to assess how well the model fits the data.


**Reasoning**:
The Latitude and Longitude columns were incorrectly converted to strings in a previous step, which will cause an error during model prediction. I need to convert them back to numeric type before making predictions and calculating metrics. After converting, I will make predictions, calculate the evaluation metrics, and print them.



```python
df['Latitude'] = df['Latitude'].astype('float')
df['Longitude'] = df['Longitude'].astype('float')

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

y_pred = lr.predict(X)

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2 = r2_score(y, y_pred)
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

print(f'R-squared: {r2:.4f}')
print(f'Mean Absolute Error (MAE): {mae:.4f}')
print(f'Mean Squared Error (MSE): {mse:.4f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
```

## Make predictions

### Subtask:
Use the trained model to make predictions on new data.


**Reasoning**:
Use the trained linear regression model to make predictions on the feature data and store the predictions in a new variable.



```python
predictions = lr.predict(X)
```

## Visualize results

### Subtask:
Create plots to visualize the model's predictions against the actual values.


**Reasoning**:
Create a scatter plot of actual vs. predicted values with a diagonal line for comparison.



```python
plt.figure(figsize=(8, 6))
plt.scatter(y, predictions, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2) # Diagonal line
plt.xlabel('Actual House Value')
plt.ylabel('Predicted House Value')
plt.title('Actual vs. Predicted House Values')
plt.grid(True)
plt.show()
```

## Interpret coefficients

### Subtask:
Examine the model's coefficients to understand the relationship between each feature and the target variable.


**Reasoning**:
Access the model coefficients and feature names, then create a DataFrame to display them together.



```python
coefficients = lr.coef_
feature_names = X.columns

coeff_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
display(coeff_df)
```
